# 1.) Load the data in the file iris.data in a 2-D array called iris

```python
import numpy as np
iris=np.loadtxt('iris.data',delimiter=',',dtype=str)
print(iris)
```

# 2.)Drop the column whose index=4 from the array iris

```python
temp=np.delete(iris,4,1)
print(temp)
```

# 3.)Display the shape, dimensions and size of iris

```python
print('Iris shape:',np.shape(iris))
print('Iris Dimensions:',iris.ndim)
print('Iris Size',iris.size)
```

# 4.)Split iris into 3 2-D arrays, each array for a different species and call them iris1,iris2,iris3

```python
idx1=iris[:,-1]=='Iris-setosa'
idx2=iris[:,-1]=='Iris-virginica'
idx3=iris[:,-1]=='Iris-versicolor'
iris=np.delete(iris,4,axis=1)
iris=iris.astype(float)
iris1=iris[idx1]
iris2=iris[idx2]
iris3=iris[idx3]
```

# 5.)Print the three arrays iris1, iris2 and iris3

```python
print('Iris 1:\n',iris1,'\n\nIris2:\n',iris2,'\n\nIris3:\n',iris3)
```

# 6.)Create a 1-D array header having elements "sepal length", "sepal width", "petal length", "petal width", "species no" in that order

```python
irisd=np.loadtxt('iris.data',delimiter=',',dtype=str)
irisd=np.array(np.insert(irisd,0,np.array(["sepal length", "sepal width", "petal length", "petal width", "species no"]),axis=0))
```

# 7.)Display the array header

```python
print(irisd)
```

# 8.)Find the max,min,mean and standard deviation for the columns of the iris  and the store the results in the arrays iris_max,iris_min,iris_avg,iris_std,iris_var respectively.The results must be rounded to not more than 2 decimal places

```python
iris=iris.astype(float)
iris_max=np.round(np.max(iris,axis=0),2)
iris_min=np.round(np.min(iris,axis=0),2)
iris_avg=np.round(np.average(iris,axis=0),2)
iris_std=np.round(np.std(iris,axis=0),2)
iris_var=np.round(np.var(iris,axis=0),2)
```

# 9.)Similarly find the max,min,mean and standard deviationfor the columns of iris1,iris2,iris3 and store the results in the arrays with appropriate names

```python
iris1_max=np.round(np.max(iris1,axis=0),2)
iris1_min=np.round(np.min(iris1,axis=0),2)
iris1_avg=np.round(np.average(iris1,axis=0),2)
iris1_std=np.round(np.std(iris1,axis=0),2)
iris1_var=np.round(np.var(iris1,axis=0),2)

iris2_max=np.round(np.max(iris2,axis=0),2)
iris2_min=np.round(np.min(iris2,axis=0),2)
iris2_avg=np.round(np.average(iris2,axis=0),2)
iris2_std=np.round(np.std(iris2,axis=0),2)
iris2_var=np.round(np.var(iris2,axis=0),2)

iris3_max=np.round(np.max(iris3,axis=0),2)
iris3_min=np.round(np.min(iris3,axis=0),2)
iris3_avg=np.round(np.average(iris3,axis=0),2)
iris3_std=np.round(np.std(iris3,axis=0),2)
iris3_var=np.round(np.var(iris3,axis=0),2)
```

# 10.)Comparisions of minimums

```python
import pandas as pd
d={
    'Iris Setosa': np.transpose(iris_min<iris1_min),
    'Iris Virginica':np.transpose(iris_min<iris2_min),
    'Iris Versicolor':np.transpose(iris_min<iris3_min)
}
df=pd.DataFrame(d)
df.index=['Sepal Length','Sepal Width','Petal Length','Petal Width']
display(df)
```
