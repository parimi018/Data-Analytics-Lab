```python
import pandas as pd
data=pd.read_csv('Mobile phone price_6thQuestion.csv')
display(data)
```

## a)	Identify the models & the price released by each brand.

```python
#data[['Brand', 'Model', 'Price ($)']]
for key, value in data.groupby(['Brand'])[['Model','Price ($)']]:
    print(key)
    display(value)
```

## b) Identify the correlation between Battery Capacity and price.

```python
lst = []
for i in data['Price ($)']:
    if ',' in i:
        i = i.replace(',', '')
    if '$' in i:
        lst.append(int(i[1:]))
    else:
        lst.append(int(i))
data['Price'] = lst
data['Battery Capacity (mAh)'].corr(data['Price'])
```

## c) Find how many models are there per each Battery capacity with same price

```python
data.groupby(['Battery Capacity (mAh)','Price ($)'])['Model'].count().reset_index()
```
